```python
import requests
from bs4 import BeautifulSoup
import time
import random
import csv
```

```python
def getHtml(url, headers):
    try:
        r = requests.get(url, headers=headers)
        r.encoding = 'utf-8'
        r.raise_for_status()
        return r.text
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None

# Fixed delay x seconds
def delay_x_0_s(fixed_delay_num):
    x = float(fixed_delay_num)
    time.sleep(x)


# Random delay 0~y seconds
def delay_0_y_s(random_delay_num):
    y = float(random_delay_num)
    time.sleep(random.random() * y)


# First fixed delay, then random delay
def delay_x_y_s(fixed_delay_num, random_delay_num):
    delay_x_0_s(fixed_delay_num)
    delay_0_y_s(random_delay_num)


# Random delay x~y seconds
def delay_between_x_y_s(start_delay_num, end_delay_num):
    x = float(start_delay_num)
    y = float(end_delay_num)
    delay_x_0_s(x)
    delay_0_y_s(y - x)

def get_ua():
    first_num = random.randint(70, 88)
    third_num = random.randint(0, 3800)
    fourth_num = random.randint(0, 140)
    os_type = [
        '(Windows NT 6.1; WOW64)', '(Windows NT 10.0; WOW64)', '(X11; Linux x86_64)',
        '(Macintosh; Intel Mac OS X 10_14_5)'
    ]
    chrome_version = 'Chrome/{}.0.{}.{}'.format(first_num, third_num, fourth_num)
    ua = ' '.join(['Mozilla/5.0', random.choice(os_type), 'AppleWebKit/537.36',
                   '(KHTML, like Gecko)', chrome_version, 'Safari/537.36']
                  )
    headers = {"user-agent": ua}
    return headers
```

```python
# Data from February to June
csv_file = "Stock_commentary_600519.csv"

with open(csv_file, mode="w", newline="", encoding="utf-8-sig") as file:
    writer = csv.writer(file)
    writer.writerow(["Read", "Reply", "Title", "Author", "Date"])
# The page number of the data viewed on the website before crawling the data
    for page in range(84, 366):
        delay_x_y_s(2, 12)
        url = "http://guba.eastmoney.com/list,600519,f_" + str(page) + ".html"
        print(url)
        headers = get_ua()
        html = getHtml(url, headers)
        if html is None:
            continue
        
        soup = BeautifulSoup(html, "html.parser")
        jobs = soup.find_all("tr", {"class": "listitem"})

        job_data = []

        for job in jobs:
            read = job.find("div", {"class": "read"}).text.strip()
            reply = job.find("div", {"class": "reply"}).text.strip()
            title = job.find("div", {"class": "title"}).text.strip()
            author = job.find("div", {"class": "author"}).text.strip()
            date = job.find("div", {"class": "update"}).text.strip()

            job_data.append([read, reply, title, author, date])

        writer.writerows(job_data)

print(f"Data has been saved to {csv_file}")
```
